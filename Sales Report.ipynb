# Exploracion de los datos


##  ¿Cuál es la tendencia general de las ventas? 

```python
# Obtencion de minimos y maximos
## Minimos
df['order_date'].min()

## Maximos
df['order_date'].max()
```

```python
# Obtencion del formato solo year and months
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
df.loc[:5, 'month_year']
```

```python
df['month_year'].loc[:5]
```

```python
# Sumar datos segun el mes
tendencia_ventas = df.groupby('month_year').sum()['sales'].reset_index()
tendencia_ventas.iloc[:5]
```

```python
# Grafico de la tendencias de ventas


## Configuracion sel grafico
plt.figure(figsize=(15,7))
plt.plot(tendencia_ventas['month_year'], tendencia_ventas['sales'])
plt.xticks(rotation='vertical',size=11)
plt.show()
```

## ¿Cuáles son los 10 productos más vendidos? (por ventas)

```python
# Ver las columnas con las que se va a trabajar

df.loc[:,['product_name','sales']]
```

```python
# Agrupando datos por meses sumando sus ventas
ventas_producto = pd.DataFrame(df.groupby('product_name').sum()['sales'])
ventas_producto
```

```python
# Ordendando los productos
ventas_producto.sort_values('sales', ascending=False)
```

```python
# Mostrando los 10 productos mas vendidos 
top_products = ventas_producto.sort_values('sales', ascending=False)
top_products[:10]
```

## ¿Cuáles son los productos más vendidos? (por cantidad)

```python
# Ver informacion deseada
df.loc[:, ['product_name', 'quantity']]
```

```python
# Agrupando por nombre del producto

most_sell = pd.DataFrame(df.groupby('product_name').sum()['quantity'])
most_sell
```

```python
# Ordenando productos por mas vendidos
most_sell_sorted = most_sell.sort_values('quantity', ascending=False)
most_sell_sorted[:10]
```

## ¿Cuál es el 'ship_mode' preferido?

```python
# Configuracion de grafico
plt.figure(figsize=(12, 9))

## Grafico
sns.countplot(data=df, x='ship_mode')
```

 ## ¿Cuáles son las categorías y subcategorías más rentables?

```python
# Preparando datos con los que trabajar
df.groupby(['category', 'sub_category']).sum()
```

```python
## De los datos solo necesitamos los beneficios (profits)
categories_profit = pd.DataFrame(df.groupby(['category', 'sub_category']).sum()['profit'])
categories_profit
```

```python
categories_profit_sorted = categories_profit.sort_values(['category', 'profit'], ascending=False)
categories_profit_sorted
```
