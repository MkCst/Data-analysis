## Uso de groupby

````python
# Generates a new dataframe grouped by subcategory 'sum'
df_by_subcategory = df.groupby('Sub-Category').sum()

## Now the subcategory is the index
df_by_subcategory.head()
````

````python
##  We can also set the index 
### Reset the index
df_by_subcategory.reset_index(inplace=True)
df_by_subcategory
````

## Definiendo funcion que agrupe automaticamente

````python
## Groupby as a function
def grouped_data(column_name):
    ''' Parametro de entrada el nombre de la columna'''
    out_df = df.groupby(column_name).sum()
    out_df.reset_index(inplace=True)
    return out_df

## Documentacion de la funcion
help(grouped_data)

## Agrupando por segmentos
grouped_data('Segment')
````
